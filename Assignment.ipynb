```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
data.head()
```

```python
HP_mean = data["HP"].mean()

def set_hp(val):
    if val < HP_mean:
        return "Low HP"
    elif val == HP_mean:
        return "Neutral HP"
    else:
        return "High HP"

speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < speed_mean:
        return "Low speed"
    elif val == speed_mean:
        return "Neutral speed"
    else:
        return "High speed"

data["High_Low_HP"] = data["HP"].apply(set_hp)
data["High_Low_Speed"] = data["Speed"].apply(set_speed)
```
