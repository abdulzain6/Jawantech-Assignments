```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/JawanTechPk/Data-Science-Course/main/netflix_titles.csv")
df
```

```python
def split_comma(data):
  data = str(data)
  if data == "nan":
    return 0
  return len(data.split(","))
```

```python
df["num_of_cast"] = df["cast"].apply(split_comma)
```
